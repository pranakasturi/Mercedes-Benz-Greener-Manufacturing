```python
#Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#Read the data from train.csv
df_train = pd.read_csv('train.csv')
```

```python
df_train.head(2)
```

```python
df_train.shape
```

```python
df_train.dtypes.head(10)
```

```python
df_train.info()
```

```python
df_train.describe()
```

```python
#Remove target as well as ID column which are not required
df_train_feature=df_train.drop(columns={'y','ID'})
```

```python
df_train_feature.shape
```

```python
train_target=df_train.y
```

```python
train_target.head()
```

```python
train_target.shape
```

```python
train_target.nunique()
```

```python
#Apply label encoder on train set
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in df_train_feature.columns:
    df_train_feature[i]=le.fit_transform(df_train_feature[i])
```

```python
df_train_feature.head()
```

```python
df_train_feature.describe()
```

```python
#Check variance on all columns of train dataframe
from sklearn.feature_selection import VarianceThreshold
var_thres = VarianceThreshold(threshold=0)
var_thres.fit(df_train_feature)
```

```python
var_thres.get_support()
```

```python
constant_columns = [column for column in df_train_feature.columns
                   if column not in df_train_feature.columns[var_thres.get_support()]]
print(len(constant_columns))
```

```python
#check column name whose variance is equal to zero
for feature in constant_columns:
    print(feature)
```

```python
#Remove columns which has,variance is equal to zero
df_train_final = df_train_feature.drop(constant_columns, axis=1)
df_train_final.head()
```

```python
df_train_final.shape
```

```python
#Check for null and unique values for train sets
df_train_final.isnull().sum()
```

```python
df_train_final.nunique()
```

```python
#Perform dimensionality reduction on train set
from sklearn.decomposition import PCA
pca = PCA(n_components=.95)
```

```python
pca.fit(df_train_final, train_target)
```

```python
df_train_final2 = pca.fit_transform(df_train_final)
print(df_train_final2.shape)
```

```python
#Predict your test_df values using XGBoost
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
```

```python
train_x,test_x,train_y,test_y = train_test_split(df_train_final2,train_target,test_size=.2,random_state=12)
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
# Fitting XGBoost to the training data
xgb_r = xgb.XGBRegressor(objective ='reg:linear', n_estimators = 10, seed = 12)
```

```python
# Fitting the model
model = xgb_r.fit(train_x, train_y)
```

```python
# Predict the model
pred = xgb_r.predict(test_x)
```

```python
pred_test_y = model.predict(test_x)
```

```python
# RMSE Computation 
rmse = np.sqrt(MSE(test_y, pred))
print("RMSE : % f" %(rmse))
```

```python
#Graph for comparing predicted values vs target values
plt.figure(figsize=(14,6))
sns.distplot(test_y[test_y<200], color="g", label="Actual value")
sns.distplot(pred_test_y[pred_test_y<200] , color="r", label="Predicted value")
plt.xlabel('Target data')
plt.title('Comparision of predicted values & target values of train dataset',weight='bold')
plt.legend(title='labels',bbox_to_anchor=(1.0, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
